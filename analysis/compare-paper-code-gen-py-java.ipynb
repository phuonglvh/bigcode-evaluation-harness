```python
import matplotlib.pyplot as plt
import numpy as np

# Data
models = ['Code Llama Python 13B', 'StarCoder 15.5B', 'CodeGeeX2 6B', 'Mistral 7B v0.1']
pass_at_1 = [29.01, 22.4, 19.0, 20.2]
pass_at_10 = [63.00, 49.8, 46.7, 50.5]
pass_at_100 = [84.01, 75.1, 68.0, 75.4]
em = [0, 0, 0, 0]
bleu = [32.6, 34.64, 33.22, 32.98]
codebleu = [24.85, 25.0, 24.83, 24.57]

x = np.arange(len(models))  # the label locations
width = 0.12  # the width of the bars

fig, ax = plt.subplots(figsize=(10, 6))

# Bar plots
rects1 = ax.bar(x - 2*width, pass_at_1, width, label='pass@1')
rects2 = ax.bar(x - width, pass_at_10, width, label='pass@10')
rects3 = ax.bar(x, pass_at_100, width, label='pass@100')
rects4 = ax.bar(x + width, em, width, label='EM')
rects5 = ax.bar(x + 2*width, bleu, width, label='BLEU')
rects6 = ax.bar(x + 3*width, codebleu, width, label='CodeBLEU')

# Add some text for labels, title and custom x-axis tick labels
ax.set_xlabel('Model')
ax.set_ylabel('Metrics')
ax.set_title('Java code generation comparison\n(t=0.8,top-p=0.95)')
ax.set_xticks(x)
ax.set_xticklabels(models)
ax.legend()

# Add values on top of bars
def add_values(rects):
    for rect in rects:
        height = rect.get_height()
        ax.annotate(f'{height:.2f}',
                    xy=(rect.get_x() + rect.get_width() / 2, height),
                    xytext=(0, 3),  # 3 points vertical offset
                    textcoords="offset points",
                    ha='center', va='bottom')

add_values(rects1)
add_values(rects2)
add_values(rects3)
add_values(rects4)
add_values(rects5)
add_values(rects6)

plt.tight_layout()
# plt.show()

plt.savefig(f"Java code generation comparison.jpg", format="jpg", dpi=1024)
```
